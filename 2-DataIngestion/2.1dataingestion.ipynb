```python
### webbase loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader("https://python.langchain.com/docs/integrations/document_loaders/")
webdocs=loader.load()
webdocs
```

```python
### webbase loader ### with getting data from specific classes
### webbase loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_path="https://lilianweng.github.io/posts/2023-06-23-agent/", 
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=("post-title","post-content")
                       ))
                       )
webdocs=loader.load()
webdocs
```

### ArxivLoader
arXiv is an open-access archive for 2 million scholarly articles in the fields of physics, mathematics, computer science, quantitative biology, quantitative finance, statistics, electrical engineering and systems science, and economics.

```python
from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(
    query="reasoning",
    # load_max_docs=2,
    # load_all_available_meta=False
)
```

```python
docs = loader.load()
print(docs[0].page_content[:100])
print(docs[0].metadata)
```
